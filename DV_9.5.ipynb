```python

from dash import Dash, dcc, html, Input, Output
import pandas as pd
import plotly.express as px
import dash_bootstrap_components as dbc
from dash_bootstrap_templates import ThemeChangerAIO, template_from_url
import plotly.graph_objects as go



dbc_css = (
    "https://cdn.jsdelivr.net/gh/AnnMarieW/dash-bootstrap-templates@V1.0.1/dbc.min.css"
)
app = Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP, dbc_css])




header = html.H4(
    " Credit score Data dashBoard", className="bg-primary text-white p-4 mb-2 text-center"
)

graph = html.Div(dcc.Graph(id="graph"), className="m-4")

dropdownData=dcc.Dropdown(
    ["dropdown_men","Payment_Amount",'Payment_Behavior','Occupation'],
    value="dropdown_men",
    id="chooseGraph",
    clearable=False,
)
dropdownDataa=dcc.Dropdown(
    ["none","Poor",'Standard','Good'],
    value="none",
    id="chooseData",
    clearable=False,
)
dropdownDatahtml=html.Div(
    children=
    [
        dbc.Label("Select graph you want to display"),
        dropdownData,
    ],
    className="mb-4",
    id='dropdownData',
)
dropdownDatahtmll=html.Div(
    children=
    [
        dbc.Label("Select data you want to display"),
        dropdownDataa,
    ],
    className="mb-4",
    id='dropdownDataa',
)
graphControl=dbc.Card(
    [dropdownDatahtml],
    body=True,
)
dataControl=dbc.Card(
    [dropdownDatahtmll],
    body=True,
)
app.layout = dbc.Container(
    [
        header,
        dbc.Row(
            [

                dbc.Col(
                    [
                        graphControl,
                        dataControl,
                        ThemeChangerAIO(aio_id="theme", radio_props={"value":dbc.themes.FLATLY}),
                    ],
                    width=2,
                ),
                dbc.Col(dcc.Graph(id='graph')),
            ]
        ),
    ],
    className="m-4 dbc",
    fluid=True,
)


@app.callback(
    Output('graph', 'figure'),
    [Input('chooseGraph', 'value'),
     Input('chooseData','value')]
)
def update_graph(datatype,dataaType):
    if dataaType=='none':
        CFPP=CFP.copy()   
    elif dataaType=='Poor':
        CFPP=CFP[CFP['Credit_Score']=='Poor']
    elif dataaType=='Standard':
        CFPP=CFP[CFP['Credit_Score']=='Standard']
    elif dataaType=='Good':
        CFPP=CFP[CFP['Credit_Score']=='Good']           

    if datatype=='dropdown_men':
        return {}
    cross_tab = pd.crosstab(CFPP.Payment_of_Min_Amount,CFPP.Credit_Score)
    if datatype=='Payment_Amount':
        crosstab=pd.crosstab(CFPP.Payment_of_Min_Amount,CFPP.Credit_Score)
        data = [
            go.Bar(
                x=cross_tab.index,
                y=cross_tab[column],
                name=column
            ) for column in cross_tab.columns
        ]
        
        layout = {
            'title': f'Cross-Tabulation Graph for {datatype}'
            
        }
        
        return {'data': data, 'layout': layout}
    
    if datatype=='Payment_Behavior':
        cross_tab = pd.crosstab(CFPP['Payment_Behaviour'], CFPP['Credit_Score'])
        
        fig = go.Figure(data=[
            go.Bar(
                x=cross_tab.index,
                y=cross_tab[column],
                name=column
            ) for column in cross_tab.columns
        ])
        
        fig.update_layout(
            title=f'Payment Behavior by Credit Score',
            xaxis_title='Payment Behavior',
            yaxis_title='Count',
            barmode='stack',
            legend_title='Credit Score'
        )
        return fig
    if datatype == 'Occupation':
        occupation_counts = CFPP['Occupation'].value_counts()
        data = [
            go.Bar(
                x=occupation_counts.index,
                y=occupation_counts.values,
            )
        ]

        layout = {
            'xaxis': {'title': 'Occupation'},
            'yaxis': {'title': 'Count'},
            'title': 'Distribution of Customers by Occupation',
            'xaxis_tickangle': -45,
        }

        return {'data': data, 'layout': layout}
        

if __name__ == "__main__":
    app.run_server(debug=False,port=8064)
```
